```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from dataset import load_gq_data
```

```python
gq_data = load_gq_data()
```

```python
gq_data[0].shape
```

```python
fontsize = 16
```

```python
fig = plt.figure(figsize=(9, 9))
plt.scatter(gq_data[0][:, 0], gq_data[0][:, 1], label="class 0")
plt.scatter(gq_data[0][gq_data[1]==1, 0], gq_data[0][gq_data[1]==1, 1], label="class 1")
plt.legend(fontsize=fontsize, loc="lower right")
plt.grid()
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.xlabel("Feature 1", fontsize=fontsize)
plt.ylabel("Feature 2", fontsize=fontsize)
plt.title("Visualization of the distribution of samples in the dataset", fontsize=fontsize)
fig.show()
```
